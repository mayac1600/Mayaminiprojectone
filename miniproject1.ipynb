```python
class GeneralNeurons:
    def __init__(self,firingRate):
        self.firingRate= firingRate

    def Activate (self, stimulus_strength):
        activation_power= stimulus_strength*self.firingRate
        return activation_power


class SensoryNeuron (GeneralNeurons):
    def __init__(self, firingRate ,receptor_type):
        super().__init__(firingRate)
        self.receptor_type= receptor_type

        def Sense_stimulus(self, stimulus_strength):
            activation_power=super().Activate(stimulus_strength)
            return activation_power
        
class Motor_neuron (GeneralNeurons):
    def __init__(self, firingRate, target_muscle):
        super().__init__(firingRate)
        self.target_muscle=target_muscle

    def control_muscle(self,target_muscle,stimulus_strength):
        target_muscle_activation=super().Activate(stimulus_strength)
        return "the muscle f{target_muscle} fires:"+target_muscle_activation
    
class Photoreceptor (SensoryNeuron):
        def __init__(self, firingRate, receptor_type):
            super().__init__(firingRate, receptor_type="light")

        def light_detection(self,light_intensity):
            firing_Rate=super.Activation(light_intensity)
            return firing_Rate
 
class Mechanoreceptor (SensoryNeuron):
        def __init__(self, firingRate, receptor_type):
            super().__init__(firingRate, receptor_type="pressure")

        def light_detection(self,pressure_intensity):
            firing_Rate=super.Activation(pressure_intensity)
            return firing_Rate
        
class Alpha_motor_neuron(Motor_neuron):
    def __init__(self, firingRate, target_muscle):
          super().__init__(firingRate, target_muscle+"skeletal muscle")
          
    def control_muscle(self,target_muscle,stimulus_strength):
        target_muscle_activation=super().Activate(stimulus_strength)
        return f"the muscle {self.target_muscle} fires:"+target_muscle_activation

class Gamma_motor_neuron(Motor_neuron):
    def __init__(self, firingRate, target_muscle):
          super().__init__(firingRate, target_muscle+"muscle spindle")
          
    def control_muscle(self,target_muscle,stimulus_strength):
        target_muscle_activation=super().Activate(stimulus_strength)
        return f"the muscle {self.target_muscle} fires:"+target_muscle_activation

                                                                                                    
```

```python
def Sin_wave(time:np.array ,frequency=5,amplitude=1):
    ### this function recives the sampling rate and for how long (from an array), the freq and amplitude and outputs the reduceds (by %20) signal

    # this is the sin func, its outputs is all the results i got from the samplling i've defined 
    signal= amplitude*np.sin(2*frequency*np.pi*time)
# creating a set of indicies the len of sampling and reduces the len by 5
    indexes_i_wanna_keep= set(np.arange(0,len(signal),5))
#the amplitudes are contants in the basic sun func, the where identifies the indecises of the amplitude and extract the relevant info from the output
    max_indexes=set( np.where(signal==amplitude)[0])
    min_indexes= set(np.where(signal==-amplitude)[0])
    # here the indcies are unioned and sorted -> creating a set of the smapled data indx
    all_indexes_i_wanna_keep= sorted(set(indexes_i_wanna_keep.union(max_indexes).union(min_indexes)))  
    #filters :)  
    downsampled_signal = signal[all_indexes_i_wanna_keep]
    downsampled_time = time[all_indexes_i_wanna_keep]

#creating a plot x,y
    plt.plot(downsampled_time,downsampled_signal)
    plt.xlabel("downsampled_time", fontsize=12)
    plt.ylabel("downsampled_signal ", fontsize=12)
    plt.title("  downsampled signal", fontsize=14)
    plt.show()
    plt.plot(time, signal, alpha=0.7)
    plt.xlabel("time", fontsize=12)
    plt.ylabel("signal ", fontsize=12)
    plt.title(" original signal", fontsize=14)

    plt.show()

time= np.arange(0,1,0.01)
Sin_wave(time)



```

```python
import re

file_path = "/Users/mayacohen/Downloads/laptop_price - dataset.csv"
data = pd.read_csv(file_path)

# Function to extract storage types 
def extract_storage_types(memory):
        # Regex to to find the storage 
        # notes  to read the syntax : d+ is to find digits (the plus is one or more) the ?:GBTB is to catch the GB/TB right after the number without diffritianting between the two 
        #the s ignores spaces and ^ is for stopping in a + 
    matches = re.findall(r"(\d+(?:GB|TB)\s*[^+]+)", memory)
    #print (matches)
    if matches:
            # Extract only the parts after "GB" or "TB" dynamically
        storage_types = []
        for match in matches:
                # Remove size to get the type. it recognisez the part with dig and tb or gb/
            type_part = re.sub(r"\d+(?:GB|TB)\s*", "", match).strip()
            storage_types.append(type_part)
        return "+".join(storage_types)

# Apply the extraction function to the "Memory" column
data["Storage type"] = data["Memory"].apply(extract_storage_types)

# Print the updated dataframe 
print(data["Storage type"].head(40))

# storing the count for each item. we will create a list of dicts here. 
storage_counts = []

# iterrows allows me to loop through each line of data
for _, row in data.iterrows():
    if "+" in row["Storage type"]:
        # Check for combined types, to prevent the type from being SDD+ HDD for exmple, 
        # Split storage types and count each one
        storage_types = row["Storage type"].split("+")
        for storage in storage_types:
            storage_counts.append({"Company": row["Company"], "Storage type": storage.strip()})
    else:
        # Directly count single storage types
        storage_counts.append({"Company": row["Company"], "Storage type": row["Storage type"].strip()})

# Convert to a DataFrame
expanded_data = pd.DataFrame(storage_counts)

# Group by Company and Storage type and count them (0 if none)
storage_distribution = expanded_data.groupby(["Company", "Storage type"]).size().unstack(fill_value=0)

# Display the result
print(storage_distribution)

# Plot the storage distribution for each company, and stacking the uses
storage_distribution.plot(kind="bar", stacked=True, figsize=(12, 6))

plt.title("Distribution of Storage Types by Company")
plt.xlabel("Company")
plt.ylabel("Count")
plt.legend(title="Storage Type")
plt.tight_layout()
plt.show()
```
